```python
import pandas as pd
import numpy as np
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
retail_dataframe = pd.read_excel("Online Retail.xlsx")
retail_dataframe.head()
```

```python
germany_retail_df = retail_dataframe[retail_dataframe['Country']=='Germany']
germany_retail_df['Description'].fillna('Desc Missing', inplace=True)
germany_retail_df['CustomerID']=germany_retail_df['CustomerID'].apply(np.int64)
```

```python
german_cancelled_trans_df = germany_retail_df[germany_retail_df['InvoiceNo'].astype('str').str.startswith('C', na=False)]
```

```python
german_sale_trans_df = germany_retail_df[~germany_retail_df['InvoiceNo'].isin(german_cancelled_trans_df.InvoiceNo)]
```

```python
german_sale_trans_df.shape[0]
```

```python
german_sale_trans_df[german_sale_trans_df['Description'].str.contains('POSTAGE')]
```

```python
def encode_data(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
def create_basket(country):
    basket = german_sale_trans_df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
    return basket 
```

```python
bk = create_basket('Germany')
basket_sets = bk.applymap(encode_data)
basket_sets.drop('POSTAGE', axis=1, inplace=True)
```

```python
freq_items = apriori(basket_sets, min_support=0.03, use_colnames=True)
freq_items['length'] = freq_items['itemsets'].apply(lambda x:len(x))
freq_items.sort_values(by='support', ascending=False).head()
```

```python
asso= association_rules(freq_items, metric='confidence', min_threshold=.65).sort_values(by='lift', ascending=False)
asso.sort_values(by="lift", ascending=False)
```

There is a 20 times more likihood to buy SPACEBOY CHILDRENS BOWL than an avg customer if they but a SPACEBOY CHILDRENS CUP. This rule is "true" in 90% of the cases (confidence). 
This can be used as insight to recommend SPACEBOY CHILDRENS BOWL for those customers who bought SPACEBOY CHILDRENS CUP.

Also, a customer is 15 times more likely to buy 'SET/6 RED SPOTTY PAPER PLATES' than an average customer (lift) if he/she buys 'SET/6 RED SPOTTY PAPER CUPS'. This rule is "true" in 87% of the cases (confidence). 
This can be used as insight to recommend SET/6 RED SPOTTY PAPER PLATES for those customers who bought SET/6 RED SPOTTY PAPER CUPS.

```python
test=german_sale_trans_df.groupby('InvoiceNo')['Description'].apply(list).reset_index()
test_desc_list=test['Description'].tolist()
test_desc_list
```

```python
german_trans = apriori(test_desc_list, min_support=0.03, min_confidence=0.65, min_lift=2, min_length=1)
```

```python
results = list(german_trans)
```

```python
values =[]
```

```python
for item in results:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    rows=(items[0],items[1],str(item[1]),str(item[2][0][2]),str(item[2][0][3]))  
    values.append(rows)
```

```python
labels=['Title1', 'Title2','Support','Confidence','Lift']
products_sugg = pd.DataFrame(data=values, columns=labels)
products_sugg['Lift']=products_sugg['Lift'].astype(float)
products_sugg.sort_values(by='Lift', ascending=False)
```
